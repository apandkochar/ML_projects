```python
X = df.drop(columns = 'medv')
Y = df['medv']
```

```python
print(len(X))
print(len(Y))
```

```python
x_train , x_test, y_train , y_test = train_test_split(X,Y,test_size=0.2 , random_state = 42)
print(x_train.shape , x_test.shape, y_train.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(x_train , y_train)
```

```python
x_train_pred = model.predict(x_train)
```

```python
x_train_acc1 = metrics.r2_score(y_train,x_train_pred )
x_train_acc2 = metrics.mean_absolute_error(y_train,x_train_pred )
```

```python
print(x_train_acc1)
print(x_train_acc2)
```

```python
x_test_pred = model.predict(x_test)
x_test_acc1 = metrics.r2_score(y_test,x_test_pred )
x_test_acc2 = metrics.mean_absolute_error(y_test,x_test_pred )
```

```python
print(x_test_acc1)
print(x_test_acc2)
```
