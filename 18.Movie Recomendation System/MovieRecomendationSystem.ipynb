```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.feature_extraction.text import TfidfVectorizer
import difflib
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("/content/drive/MyDrive/AI/YT Machine learning projects/18.Movie Recomendation System/movies.csv")
data.head()
```

```python
data.info()
```

```python
imp_feature = ['genres' , 'keywords', 'tagline' , 'cast' , 'director' , 'overview']
for i in imp_feature:
  data[i] = data[i].fillna('')
```

```python
combined_feature = data['genres'] + ' ' + data['keywords'] + ' ' + data['tagline'] + ' ' + data['cast']+' ' + data['director']+' ' + data['overview']
combined_feature
```

```python
vectorizer = TfidfVectorizer()
featrue_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(featrue_vector)
```

```python
similarity = cosine_similarity(featrue_vector)
similarity
```

```python
movie_input = input("Enter your fav. movies that you want to watch:")
movie_input
```

```python
list_of_all_movies = data['title'].tolist()
```

```python
close_matchs = difflib.get_close_matches(movie_input , list_of_all_movies)
close_matchs
```

```python
close_match = close_matchs[0]
close_match
```

```python
index_of_movie = data[data.title == close_match]['index'].values[0]
index_of_movie
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
```

```python
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True )
sorted_similar_movies[:30]
```

```python
movie = 0
for i in sorted_similar_movies:
  index = i[0]
  title_of_movie = data[data.index == index]['title'].values[0]
  if (movie < 30):
    print(movie , '.' , title_of_movie)
    movie += 1
```

```python
movie_input = input("Enter your fav. movies that you want to watch:")

list_of_all_movies = data['title'].tolist()

close_matchs = difflib.get_close_matches(movie_input , list_of_all_movies)

close_match = close_matchs[0]

index_of_movie = data[data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True )

movie = 0
for i in sorted_similar_movies:
  index = i[0]
  title_of_movie = data[data.index == index]['title'].values[0]
  if (movie < 30):
    print(movie , '.' , title_of_movie)
    movie += 1
```
